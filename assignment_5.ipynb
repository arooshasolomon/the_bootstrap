2. Imagine rolling three dice, $d_1$, $d_2$, and $d_3$. Let $R_{med}$ be the middle of the three face values. So if you roll 2, 3 and 4, the middle value is 3, and if you roll 2, 4, 4, the middle value is 4, and so on. What is the probability of getting a 1, 2, 3, 4, 5, or 6? 3. Imagine rolling a die. If you roll 1, 2, 3, 4, or 5, add that number to your total and stop; if you roll a six, add that number to your total and roll the die again. So you could roll, say, two sixes and then a four, and get a total of 16, or one three and get a total of 3, or twelve sixes and 1 and get 72, and so on. Write code to simulate this process, and determine its expected value using the law of large numbers. What is the probability of getting a total of 1, 2, 3, ... and so on, in your simulation? I recommend using simulations and the law of large numbers.  (Hint: The `while` loop might be useful in this case.)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

## 1.
faces = np.array([1,2,3,4,5,6])

# Explicit calculation:
pdf = np.zeros(6)
for i in range(6):
    for j in range(6):
        min_ij = np.min([faces[i],faces[j]])
        pdf[min_ij-1] += 1
pdf = pdf/36
cdf = np.cumsum(pdf)

# Bootstrap simulation:
N = 1000
d1 = np.random.choice(faces, size=N)
d2 = np.random.choice(faces, size=N)
min_roll = [ min([d1[i], d2[i]]) for i in range(N) ]

ecdf = np.zeros(6)
for i in range(6):
    ecdf[i] = (1/N)*np.sum( min_roll <= faces[i] )

# Comparison:
plt.scatter(faces,ecdf,label='ECDF')
plt.plot(faces,ecdf,label='ECDF')
plt.scatter(faces,cdf,label='CDF')
plt.plot(faces,cdf,label='CDF')
plt.xlabel("x")
plt.ylabel("Probability")
plt.legend(loc='lower right')
plt.title('Empirical and Theoretical CDF')
plt.show()
```

```python
## 3.

# Define the number of trials and the faces of the die
N = 50000
faces = np.array([1, 2, 3, 4, 5, 6])

# Create an array to store the results
results = []

# Simulate the rolling process
for _ in range(N):
    total_sum = 0
    while True:
        roll = np.random.choice(faces)
        total_sum += roll
        if roll < 6:
            results.append(total_sum)
            break

# Convert results to a DataFrame for easier analysis
results = pd.DataFrame(results)

# Print descriptive statistics
print(results.describe())

# Plot histogram and ECDF
sns.histplot(results, kde=True)
plt.title("Histogram of Total Sum")
plt.show()

sns.ecdfplot(results)
plt.title("ECDF of Total Sum")
plt.show()
```

**Q2.** This question refers to the `mammogram.csv` data. It has two variables, `treatment` which takes the values `control` or `mammogram`, and `breast_cancer_death`, which takes the values `no` or `yes`. This is an experiment that followed 89,835 women for 25 years to see if mammograms were superior to more traditional breast cancer screenings in preventing breast cancer deaths.

1. Cross tabulate `treatment` and `breast_cancer_death`. What is the difference in 25-year survival rates between the control and mammogram groups?
2. Bootstrap the densities and distributions of survival rates for the two groups.
3. Construct a 99% confidence interval for the difference in outcomes bewteen the two groups. Does it include zero?
4. We're not doctors, these were just some intriguing data, and the information about the patients is extremely sparse. Why might these data over/understate the conclusions you've reached? What other data would you like to have to better understand or criticize your results?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('mammogram.csv')

## 1. Cross-tabulation and raw treatment effect

# Cross-tabulation of treatment vs. breast cancer death
ctab = pd.crosstab(df['treatment'], df['breast_cancer_death'])
print(ctab, '\n')

# Cross-tabulation with margins (normalized)
ctab_norm = pd.crosstab(df['treatment'], df['breast_cancer_death'], margins=True, normalize='all')
print(ctab_norm, '\n')

# Calculate the raw treatment effect manually
raw_treatment_effect = (44425 / 44925) - (44405 / 44910)
print(f'Raw treatment effect: {raw_treatment_effect:.4f}')

## 2. Resampling for treatment effect

# Convert 'breast_cancer_death' to a binary 'survive' column
df['survive'] = np.where(df['breast_cancer_death'] == 'no', 1, 0)

# Filter data for treatment and control groups
df_treat = df[df['treatment'] == 'mammogram']
df_control = df[df['treatment'] == 'control']

# Set number of bootstrap samples
num_samples = 5000
treatment_effects = []

# Bootstrap resampling loop to compute treatment effects
for _ in range(num_samples):
    # Bootstrap sample for treatment and control groups
    boot_treat = df_treat.sample(n=len(df_treat), replace=True)
    boot_control = df_control.sample(n=len(df_control), replace=True)

    # Compute treatment effect for the bootstrap sample
    treatment_effect = boot_treat['survive'].mean() - boot_control['survive'].mean()
    treatment_effects.append(treatment_effect)

# Plot Kernel Density Estimate (KDE) of the treatment effects
sns.kdeplot(treatment_effects, fill=True, color='blue')
plt.title('KDE of Treatment Effect')
plt.show()

# Plot Empirical CDF of the treatment effects
sns.ecdfplot(treatment_effects, color='green')
plt.title('ECDF of Treatment Effect')
plt.show()

## 3. Confidence Interval Calculation

# Calculate the 99% confidence interval from the bootstrap distribution
lower_bound = np.percentile(treatment_effects, 0.5)
upper_bound = np.percentile(treatment_effects, 99.5)

print(f'99% confidence interval for treatment effect: ({lower_bound:.4f}, {upper_bound:.4f})')

#sourced from chat
```

The 99% confidence interval includes zero, indicating that undergoing a mammogram does not lead to a statistically significant difference in 25-year survival rates.
